```python
def get_dis_and_costs(x, what='distancia'):
    r = re.search('((?<=>)\d+\,?\d+\.?\d+)(\W+)', x)
    r2 = re.search('((?<=>)\d+):(\d+)', x)
    if what == 'distancia' or what == 'casetas' or what == 'combustible' or what == 'total':
        if not r:           #si r no tiene sentido
            return np.nan
        return float(r.group(1).replace(',',''))
    elif what == 'tiempo':
        if not r2:
            return np.nan
        h = float(r2.group(1)) # convertimos a float para poder operar
        m = float(r2.group(2))
        b= h + m * 1 / 60
        return b  
```

```python
distancias['distancia'] = distancias['distancia'].apply(get_dis_and_costs, what='distancia') 
distancias['cost_casetas'] = distancias['cost_casetas'].apply(get_dis_and_costs, what='casetas') 
distancias['cost_combustible'] = distancias['cost_combustible'].apply(get_dis_and_costs, what='casetas') 
distancias['cost_total'] = distancias['cost_total'].apply(get_dis_and_costs, what='casetas') 
distancias['tiempo'] = distancias['tiempo'].apply(get_dis_and_costs, what='tiempo') 

```
